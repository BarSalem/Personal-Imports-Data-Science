```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from scipy.stats import chi2_contingency
import seaborn as sns
from collections import Counter
from sklearn.decomposition import PCA
df2=pd.read_csv("fileSSSS.csv")
```

```python
df=df2.copy()
```

```python
Brand=df["brand"].unique()
```

```python
for i in range(len(Brand)):
    if(Brand[i] in df.brand.values):
        df.loc[df["brand"]==Brand[i],"brand"] = int(i+1)
```

```python
typeofVeicle=df["typeofVeicle"].unique()
for i in range(len(typeofVeicle)):
    if(typeofVeicle[i] in df.typeofVeicle.values):
        df.loc[df["typeofVeicle"]==typeofVeicle[i],"typeofVeicle"] = int(i+1)
```

```python
Gasoline=df["Gasoline"].unique()
for i in range(len(Gasoline)):
    if(Gasoline[i] in df.Gasoline.values):
        df.loc[df["Gasoline"]==Gasoline[i],"Gasoline"] = int(i+1)
```

```python
df.drop(['Unnamed: 0.1.1.1.1.1.1'], axis=1,inplace=True)
```

```python
df = df.rename(columns={'typeofVeicle': 'Brand' , "brand":"Type of Vehicle","country":"Country","Gasoline":"Engine Type"})
```

```python
bins=[1000,2000,3000,4000,5000,6000,10000]
labels=[1,2,3,4,5,6]
df['EngineSize'] = pd.cut(df['EngineSize'], bins=bins, labels=labels)
```

```python
dftry=df.copy()
```

```python
outliers=["Country","Brand","Type of Vehicle"]
```

```python
for col in outliers:
    if(dftry[col].dtype!=object):
        Q1=np.percentile(dftry[col],15)
        Q3=np.percentile(dftry[col],85)
        IQR=Q3-Q1
        IQR_Range=1.5*IQR
        dftry[col][(dftry[col]< Q1-IQR_Range) | (dftry[col] > Q3+IQR_Range)]=np.nan
```
